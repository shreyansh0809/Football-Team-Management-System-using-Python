```python
#Football Team Management System
global playerlist
playerlist = ["Cristiano Ronaldo", "Sergio Ramos", "Karim Benzema", "Gareth Bale"]  #some players already exist

def playermanagement():
    #Ask the choice from user
    print("\n⚽⚽⚽⚽⚽⚽⚽ Welcome to Real Madrid Player Management System ⚽⚽⚽⚽⚽⚽⚽\n")
    print("[Choice 1: Show the List of Players]")
    print("[Choice 2: Add New Player]")
    print("[Choice 3: Search for Player]")
    print("[Choice 4: Delete a Player]\n")

    try:
        x = int(input("Enter a choice: "))
    except ValueError:
        exit("\This is not a Number! Enter a Valid Number")
    else:
        print("\n")
    #Show the Player List of the Team 
    if(x==1):
        print("Player List\n")
        for players in playerlist:
            print("++ {} ++".format(players))
    #Add a player to the team
    elif(x==2):
        playernew = input("Enter New Player: ")
        if(playernew in playerlist):
            print("\nThis Player {} is already in the Team".format(playernew))
        else:
            playerlist.append(playernew)
            print("\n++ New Player {} Added Successfully ++\n".format(playernew))
        for players in playerlist:
            print("++ {} ++".format(players))
    #Search for the player whether he exists in the team
    elif(x==3):
        playersearching = input("Enter the Player Name to Search: ")
        if(playersearching in playerlist):
            print("\n++ There is a Record Found of the Player {} ++".format(playersearching))
        else:
            print("\n++ There is No Record Found Of the Player {} ++".format(playersearching))
    #Delete the player from team
    elif(x==4):
        playerdelete = input("Choose a Player Name To Delete: ")
        if(playerdelete in playerlist):
            playerlist.remove(playerdelete)
        for players in playerlist:
            print("++ {} ++".format(players))
        else:
            print("\n++ There is No Record Found of the Player in this Team{} ++".format(playerdelete))
    else:
        print("Enter Valid Choice!!")
        


playermanagement()

def continueAgain():
    runningagain = input("\nDo you want to continue the process Yes/No?: ")
    if(runningagain.lower() == 'yes'):
        playermanagement()
        continueAgain()
    else:
        quit()

continueAgain()
```
